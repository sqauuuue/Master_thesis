```python
import numpy as np


# Calculate the IQR
q25_fc_ec = np.percentile(data_array_fc_ec_values, 25, axis=1, keepdims=True)
q75_fc_ec = np.percentile(data_array_fc_ec_values, 75, axis=1, keepdims=True)
iqr_fc_ec = q75_fc_ec - q25_fc_ec

# Define lower and upper bounds for identifying outliers(1-99)
lower_bound_fc_ec = q25_fc_ec - 1.5 * iqr_fc_ec
upper_bound_fc_ec = q75_fc_ec + 1.5 * iqr_fc_ec


lower_bound_fc_ec



#Identify outliers

outliers = (data_array_fc_ec_values < lower_bound_fc_ec) | (data_array_fc_ec_values > upper_bound_fc_ec)
outliers_fc_ec = np.array(outliers)
data_array_fc_ec_values_array = np.array(data_array_fc_ec_values)




outliers_fc_ec_1 = data_array_fc_ec_values_array[outliers_fc_ec]
outliers_fc_ec_1
```
