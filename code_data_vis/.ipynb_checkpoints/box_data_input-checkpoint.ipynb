```python
mean_ds_ec
```

```python
start_index = 56
step_size = 1
num_iterations =15
fc_ec_values=[]
for i in range(num_iterations):
    fc_ec_value = mean_ds_ec[i * step_size,:, start_index - i * step_size*4]
    fc_ec_values.append(fc_ec_value)
data_array_fc_ec_values =xr.DataArray(fc_ec_values)
```

```python
data_array_fc_ec_values_trans = data_array_fc_ec_values.T
bp = plt.boxplot(data_array_fc_ec_values_trans,whis=[5, 95], showfliers= False)
```
