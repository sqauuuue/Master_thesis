```python
import numpy as np
%store -r data_array_ec_values
def divergence_function(f, g):
    M = 50
    N = 50
    
    divergence_sum = 0
    divergence_sum_1 = 0
    divergence_sum_2 = 0
    
    for i in range(M):
        for j in range(N):
            divergence_sum += (f[i] - g[j])
            divergence_sum_1 += (f[i] - f[j])
            divergence_sum_2 += (g[i] - g[j])
            
    
    divergence = (1 / (M * N)) * divergence_sum - (1 / (2 * M * M)) * divergence_sum_1 - (1 / (2 * N * N)) * divergence_sum_2
    
    return divergence


```
